```python
def render_latex_table_simple(sorted_df, increasing_metrics, res_b4_training, 
                               metric_name_map=None, model_name='gpt2', 
                               decimals=1, decimals_average=2, output_path=None):
    """
    Generate LaTeX table from sorted_df showing improvements over baseline.
    
    Args:
        sorted_df: DataFrame with rows sorted by performance
        increasing_metrics: List of metric column names to include (where higher is better)
        res_b4_training: DataFrame with initial (before training) baseline values for all model sizes
        metric_name_map: Dict mapping metric column names to short names (e.g., {'anli_r2,acc,none': 'ANLI-R2'})
        model_name: Name of the model being trained (e.g., 'gpt2') - used to determine baseline
        decimals: Number of decimal places for metric values
        decimals_average: Number of decimal places for average column
        output_path: Path to save the .tex file (default: ./tables/results_table.tex)
    """
    import os
    import numpy as np
    import pandas as pd
    
    # Build rows list
    rows = []
    
    # Rows 1-4: All pretrained models (no mid-training) - from res_b4_training
    for _, model_row in res_b4_training.iterrows():
        model = model_row['model']
        rows.append({
            "method": f"Pretrained ({model})",
            "src": "initial",
            "data": model_row
        })
    
    # Get baseline reference (the specific model being trained)
    baseline_reference = res_b4_training[res_b4_training['model'] == model_name].iloc[0]
    
    # Row 5: Default loss (None-1 baseline best)
    baseline_row = sorted_df[sorted_df['disp'] == 'None-1'].iloc[0]
    rows.append({
        "method": "Default loss",
        "src": "baseline_best",
        "data": baseline_row
    })
    
    # Remaining rows: dispersion methods (excluding None-1)
    for _, row in sorted_df[sorted_df['disp'] != 'None-1'].iterrows():
        method = row['disp_type'] if pd.notna(row['disp_type']) else "None"
        rows.append({
            "method": method,
            "src": "run",
            "data": row
        })
    
    # Build table header - use metric_name_map if provided
    if metric_name_map is None:
        header_names = [m.split(',')[0].replace('_', r'\_') for m in increasing_metrics]
    else:
        header_names = [metric_name_map.get(m, m.split(',')[0]).replace('_', r'\_') 
                        for m in increasing_metrics]
    header_names.append("Average")
    
    column_alignment = "l " + " ".join(["c"] * len(header_names))
    
    lines = []
    lines.append(r"\begin{tabular}{" + column_alignment + r"}")
    lines.append(r"\toprule")
    lines.append("Method & " + " & ".join(header_names) + r" \\")
    lines.append(r"\midrule")
    
    # Process each row
    for row_info in rows:
        # Escape underscores in method name
        method_escaped = row_info['method'].replace('_', r'\_')
        metric_cells = []
        values_for_average = []
        baseline_values_for_average = []
        
        for metric_name in increasing_metrics:
            # Get value and baseline
            value = row_info['data'][metric_name]
            baseline_value = baseline_reference[metric_name]
            
            if not np.isfinite(value):
                metric_cells.append("N/A")
                continue
            
            cell_text = f"{value:.{decimals}f}"
            
            # Add colored difference (except for initial rows)
            if row_info['src'] != "initial" and np.isfinite(baseline_value):
                difference = np.round(value, decimals) - np.round(baseline_value, decimals)
                sign = "+" if difference >= 0 else ""
                color_name = "forestgreen" if difference >= 0 else "crimson"
                cell_text += f"$_{{\\textcolor{{{color_name}}}{{({sign}{difference:.{decimals}f})}}}}$"
            
            metric_cells.append(cell_text)
            if np.isfinite(value):
                values_for_average.append(value)
            if np.isfinite(baseline_value):
                baseline_values_for_average.append(baseline_value)
        
        # Calculate average
        if values_for_average:
            average_value = float(np.mean(values_for_average))
            average_baseline = float(np.mean(baseline_values_for_average)) if baseline_values_for_average else np.nan
            
            if row_info['src'] == "initial" or not np.isfinite(average_baseline):
                average_cell_text = f"{average_value:.{decimals_average}f}"
            else:
                difference = np.round(average_value, decimals_average) - np.round(average_baseline, decimals_average)
                sign = "+" if difference >= 0 else ""
                color_name = "forestgreen" if difference >= 0 else "crimson"
                average_cell_text = f"{average_value:.{decimals_average}f}$_{{\\textcolor{{{color_name}}}{{({sign}{difference:.{decimals_average}f})}}}}$"
        else:
            average_cell_text = "N/A"
        
        metric_cells.append(average_cell_text)
        lines.append(method_escaped + " & " + " & ".join(metric_cells) + r" \\")
    
    lines.append(r"\bottomrule")
    lines.append(r"\end{tabular}")
    
    # Save and print
    table_tex = "\n".join(lines)
    if output_path is None:
        os.makedirs("./tables", exist_ok=True)
        output_path = "./tables/results_table.tex"
    
    with open(output_path, "w") as f:
        f.write(table_tex)
    
    print("\n===== LaTeX table =====\n")
    print(table_tex)
    print(f"\n[Saved to: {output_path}]\n")
    
    return table_tex
```
